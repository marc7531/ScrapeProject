```python
#PROCESS FOR THE ENGINE COLUMN = SPLIT BY WHITESPACE
def enginesplit(eng):
    outputlist = eng.split(' ')
    liter = outputlist[0]
    cylinder = outputlist[1]

#PROCESS TO MODIFY THE TIRES COLUMN
def cleantire(a_tire):
    tiresplit = re.findall('\d+',a_tire)
    tirenew = str(tiresplit[0])+'-'+str(tiresplit[1])
    return(tirenew)

def cleanmpg(mpgvalue):
    mpgsplit = re.findall('\d+',mpgvalue)
    mpgsplit = int(mpgsplit[0])
    return(mpgsplit)

#PROCESS HORSEPOWER
def cleanhorse(hp):
    hpsplit = re.findall('\d+',hp)
    return(hpsplit)

#PROCESS TO CLEAN MILEAGE AND PRICE 
def cleanmilesprice(milesorprice):
    cleanvar = re.findall('\d+',milesorprice)
    cleancat = ''
    for i in cleanvar:      
        cleancat = cleancat+i
    cleancat = int(cleancat)
    return(cleancat)
```

```python
#PROCESS FOR TRANSMISSION COLUMN
def transspeed(trans):
    speed = re.findall('\d+',trans)
    return(speed)

def transtype(trans):
    ttype=''
    match = re.search('auto',trans)
    match1 = re.search('Auto',trans)
    match2 = re.search('manual',trans)
    match3 = re.search('Manual',trans)
    if match or match1:
        ttype = 'auto'
    if match2 or match3:
        ttype = 'manual'
    return(ttype)
```

```python
#THIS WILL SPLIT THE MAKE AND MODEL INTO TWO COLUMNS !!
# new data frame with split value columns 
new = df2["Make Model"].str.split(" ", n = 1, expand = True) 
  
# making seperate first name column from new data frame 
df2["Brand"]= new[0] 
  
# making seperate last name column from new data frame 
df2["Model"]= new[1] 
#============================================================================================
#THIS WILL SPLIT THE ENGINE INTO TWO COLUMNS !!
# new data frame with split value columns 
newengine = df2["Engine"].str.split(" ", n = 1, expand = True) 
  
# making seperate first name column from new data frame 
df2["Liters"]= newengine[0] 
  
# making seperate last name column from new data frame 
df2["Cylinder"]= newengine[1] 
df2.head()
```
