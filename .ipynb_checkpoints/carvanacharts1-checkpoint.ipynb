```python
f = open('carvanascrapedata.csv')
cardata = pd.read_csv('carvanascrapedata.csv')
f.close()
```

```python
cardata.columns
```

```python
sns.violinplot(x="Year", y="Price", data=cardata)
plt.ylim(5000,50000)
plt.title("Relationship between Price and Age of Vehicle")
```

```python
sns.violinplot(x="Year", y="Mileage", data=cardata)
plt.title("Range of Mileage by Year")
```

```python

a = sns.distplot(cardata.Price,bins=100)
plt.title("Distribution of Prices in Inventory Sample")
plt.xlim(10000,40000)
```

```python
brandprice = cardata.groupby('Brand').size().plot.barh()
brandprice.set_title("Number of Cars per Brand")
brandprice2 = cardata.groupby('Brand').agg({'Price':'mean'}).sort_values('Price').plot.barh()
brandprice2.set_title("Average Price Point per Brand")
#brandprice.plot.barh()
#brandprice2.plot.barh()
```

```python
relationtoprice = cardata.filter(['Price','Mileage','HP','Highway','Cylinder'])
sns.pairplot(relationtoprice)
```

```python
df4 = cardata.groupby('Liters').filter(lambda x: len(x)>30)
#df4 = df4.groupby('Cylinder').filter(lambda x:len(x)>2)
literlist = ['1.6L','1.8L','2.0L','2.4L','2.5L','3.5L','3.6L','Pentastar','Ecotec']
literplot = sns.barplot(x=df4['Liters'], y=df4['Price'],order=literlist)
literplot.set_title("Engine Liters and Price")
```

```python
enginetype = sns.barplot(x=df4['Liters'],y=df4['HP'],order=literlist)
enginetype.set_title("Engine Liters and Horsepower")
```
