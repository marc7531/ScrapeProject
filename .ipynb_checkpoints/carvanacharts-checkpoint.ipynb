```python
#Data has been combined & cleaned in a separate Jupyter Notebook
f = open('carvanascrapedata.csv')
cardata = pd.read_csv('carvanascrapedata.csv')
f.close()
```

```python
import qgrid
qgrid_widget = qgrid.show_grid(cardata,show_toolbar=True)
qgrid_widget
```

```python
cardata.VIN.head(10)
```

```python
d = sns.violinplot(x="Year", y="Price", data=cardata)
plt.ylim(5000,50000)
plt.title('Relationship between Price and Age of Vehicle')
```

```python
e = sns.violinplot(x="Year", y="Mileage", data=cardata)
plt.title("Relationship between Mileage and Age")
```

```python

a= sns.kdeplot(cardata.Price, shade=True, label='PDF of Price')
plt.xlim(10000,40000)
plt.title("Probability Density of Vehicle Prices")
```

```python
brandprice = cardata.groupby('Brand').size().sort_values().plot.barh()
brandprice.set_title("Number of Cars per Brand")
```

```python
brandprice2 = cardata.groupby('Brand').agg({'Price':'mean'}).sort_values('Price').plot.barh()
brandprice2.set_title("Average Price Point per Brand")
```

```python
fuelmask = ((cardata['Fuel']!='Gas')&(cardata['Fuel']!='Diesel'))
bf = cardata[fuelmask].sort_values('count')

fuelcounts = sns.countplot(y="Brand", hue="Fuel",data=bf,dodge=False)
plt.suptitle = 'Brands offering Fuel Efficient Vehicles'
```

```python
relationtoprice = cardata.filter(['Price','Mileage','HP','Highway','Cylinder'])
sns.pairplot(relationtoprice)
```

```python
df5 = cardata.filter(['Price','Mileage','HP','Highway','City','Passengers'])
highwaymask = df5.Highway<45
df6 = df5[highwaymask]
corr = df6.corr()

# plot the heatmap
corry=sns.heatmap(corr, xticklabels=corr.columns,yticklabels=corr.columns)
corry.set_title("Correlation Heatmap")
```

```python
df7 = cardata.filter(['Price','Highway','Fuel'])
df7 = df7[highwaymask]
sns.lmplot(x="Highway",y="Price",data=df7,hue="Fuel", palette="Set2")  
plt.xlim(15,45)
plt.ylim(0,40000)
plt.xlabel("Highway MPG")

```

```python
sns.regplot(x="HP",y="Price",data=df5)
```

```python
df4 = cardata.groupby('Liters').filter(lambda x: len(x)>30)
#df4 = df4.groupby('Cylinder').filter(lambda x:len(x)>2)
literlist = ['1.6L','1.8L','2.0L','2.4L','2.5L','3.5L','3.6L','Pentastar','Ecotec']
literplot = sns.barplot(x=df4['Liters'], y=df4['Price'],order=literlist)
literplot.set_title("Engine Liters and Price")
```

```python
enginetype = sns.barplot(x=df4['Liters'],y=df4['HP'],order=literlist)
enginetype.set_title("Engine Liters and Horsepower")
```

```python
toptire = df4.Tires.value_counts().head(5)
tireorder = ['215-55','225-60','225-65','235-45','235-55']

dataOfTopTires = df4[df4['Tires'].map(lambda x: x in toptire)]

sns.boxplot(x="Tires",y="HP",data=dataOfTopTires,order=tireorder)
#plt.ylabel("Horsepower")
```

```python
toptire1 = df4.Tires.value_counts().head(8)
dataOfTopTires1 = df4[df4['Tires'].map(lambda x: x in toptire1)]

sns.stripplot(x='Tires',y='Price',data=dataOfTopTires1)
```

```python
df4['Exterior'] = df4['Exterior'].str.lower()
topextcolor = df4.Exterior.value_counts().head(8)
extorder = ['black','charcoal','gray','beige','jet black','ebony','graphite','almond']
dataOfTopColors = df4[df4['Exterior'].map(lambda x: x in topextcolor)]
colorcounts = sns.countplot(x="Exterior", data=dataOfTopColors,order=extorder)
plt.title("Most Represented Exterior Vehicle Colors")
```

```python
dataOfTopColors = df4[df4['Exterior'].map(lambda x: x in topextcolor)]
colorprice = sns.stripplot(x="Exterior", y="Price",data=dataOfTopColors,order=extorder)
brandOfTopColor = dataOfTopColors.groupby('Exterior').agg({'Price':'median'}).sort_values('Price',ascending=False)
brandOfTopColor
```

```python
transcounts = sns.countplot(x="TransType",data=cardata)
plt.title("Vast Majority of Vehicles with Automatic Transmissions")
```
